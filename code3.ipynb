Splitting the data

```python
X_tr_data = train_data.iloc[:,0:41]
y_tr_data = train_data.iloc[:,41]

X_ts_ukn_data = test_unknown_data.iloc[:,0:41]
y_ts_ukn_data = test_unknown_data.iloc[:,41]

X_ts_kn_data = test_known_data.iloc[:,0:41]
y_ts_kn_data = test_known_data.iloc[:,41]
```

```python
scaler = Normalizer()
scaler.fit(X_tr_data)
scaled_train_data=scaler.transform(X_tr_data)
scaled_test_unknown_data=scaler.transform(X_ts_ukn_data)
scaled_test_known_data=scaler.transform(X_ts_kn_data)
scaler
```

```python
p_c_a = PCA(n_components = 28)#principal component analysis object creation
p_c_a.fit(scaled_train_data) #fitting
```

```python
x_train_pca = p_c_a.transform(scaled_train_data) #transforming
x_test_unknown_pca = p_c_a.transform(scaled_test_unknown_data) #transforming
x_test_known_pca = p_c_a.transform(scaled_test_known_data) #transforming
```

```python
print(x_train_pca.shape)  # data set after performing p_c_a on known data set
print(x_test_known_pca.shape) # data set after performing p_c_a on known data set
print(x_test_unknown_pca.shape) # data set after performing pca on unknown data set
```

```python
def show_confusion_matrix(y_tst,y_prd,tt):
    # A function that draws the confusion matrix of the model and the data
    cm_matrix = confusion_matrix(y_tst,y_prd)
    # Creating a dataframe so it will be easy for plotting.
    cm = pnd.DataFrame(cm_matrix,
                     index = ['Normal','Dos','Probe','R2l','U2r'],
                     columns = ['Normal','Dos','Probe','R2l','U2r'])
    #Plotting the confusion matrix
    palt.figure(figsize=(10,6))
    seab.heatmap(cm,cmap=palt.cm.Blues, annot=True,fmt='d')
    palt.title(tt)
    palt.ylabel('Actal Values')
    palt.xlabel('Predicted Values')
    palt.show()
    return cm_matrix                                                                                                         # Prediction with the help of the model
```

```python
#detection rate calculation
def det_rate_calc(cm):
    F_P = cm.sum(axis=0) - nmp.diag(cm)
    F_N = cm.sum(axis=1) - nmp.diag(cm)
    T_P = nmp.diag(cm)
    T_N = cm.sum() - (F_P + F_N + T_P)

    # Sensitivity, hit rate, detection rate, or true positive rate
    drate = T_P/(T_P+F_N)
    drate=drate.mean()
    return drate
```

Applying Light GBM classifier On Training data

```python
%%time
# we define classifier with hyperparameter.
lg = LGBMClassifier(learning_rate=0.05, max_depth=20,n_estimators=50)
lg.fit(x_train_pca,y_tr_data) # fitting the lgbm on the dataset
tr_scores2 = cross_val_score(lg, x_train_pca,y_tr_data, cv=5).mean()
print("Accuracy after performing LGBM (train data)",tr_scores2)
print()
# train known attacks
y_forecast_tr_lg = lg.predict(x_train_pca)
print("Classification Report for Train Dataset")
print(classification_report(y_tr_data,y_forecast_tr_lg,labels=[0,1,2,3,4]))
print()
tr_matrix3 = show_confusion_matrix(y_tr_data,y_forecast_tr_lg,'Train LGBM Confusion Matrix')
print()
print('Detection Rate of LGBM (Train Dataset) :',det_rate_calc(tr_matrix3))
```

```python
%%time
ts_kn_scores2 = cross_val_score(lg, x_test_known_pca,y_ts_kn_data, cv=5).mean()
print("Accuracy after performing LGBM (test known data)",ts_kn_scores2)
print()
# test known attacks
y_forecast_kn_lg = lg.predict(x_test_known_pca)
print("Classification Report for Known Test Dataset")
print(classification_report(y_ts_kn_data,y_forecast_kn_lg,labels=[0,1,2,3,4]))
print()
kn_matrix3 = show_confusion_matrix(y_ts_kn_data,y_forecast_kn_lg,'Known LGBM Confusion Matrix')
print()
print('Detection Rate of LGBM (Known Dataset) :',det_rate_calc(kn_matrix3))
```

```python
%%time
ts_ukn_scores2 = cross_val_score(lg, x_test_unknown_pca,y_ts_ukn_data, cv=5).mean()
print("Accuracy after performing LGBM (test unknown data)",ts_ukn_scores2)
print()
# test unknown attacks
y_forecast_ukn_lg = lg.predict(x_test_unknown_pca)
print("Classification Report for Unknown Test Dataset")
print(classification_report(y_ts_ukn_data,y_forecast_ukn_lg,labels=[0,1,2,3,4]))
ukn_matrix3 = show_confusion_matrix(y_ts_ukn_data,y_forecast_ukn_lg,'Unknown LGBM Confusion Matrix')
print()
#detection rate
print('Detection Rate of LGBM (Unknown Test Dataset) :',det_rate_calc(ukn_matrix3))
```
